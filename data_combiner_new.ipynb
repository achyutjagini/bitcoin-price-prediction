```python
#x â€² = ( x âˆ’ x m i n ) / ( x m a x âˆ’ x m i n )
def normalizeCol(col):
  col=list(col)
  maxx=max(col)
  mini=min(col)
  for i in range(len(col)):
    col[i]=(col[i]-mini)/(maxx-mini)
  return col
```

```python
#score is date:[volume,avg score]
score={}
user_followers={}

for i in range(len(tweet_data.user_profession)):
  if tweet_data.date[i] in score:
    score[tweet_data.date[i]][0]+=1 #volume of tweets
    score[tweet_data.date[i]][1]+=tweet_data.score[i] #sum of tweet scores for that day
    
    

  else:
    score[tweet_data.date[i]]=[1,tweet_data.score[i]]
    

#taking avg
for i in score:
  score[i][1]=score[i][1]/score[i][0]  # average tweet score for that day


```
