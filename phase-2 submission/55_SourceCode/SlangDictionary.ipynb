```python
def addToDict(slang,meaning):
  try:
    SLANG_DICT[slang]
  except:
    SLANG_DICT[slang]=meaning
```

##"https://www.webopedia.com/reference/text-abbreviations/"

```python
URL = "https://www.webopedia.com/reference/text-abbreviations/"
#1500+ slangs

page = requests.get(URL,headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
c = soup.find(class_='row-hover')                             
slangs=c.find_all(class_="column-1")
meanings=c.find_all(class_="column-2")
for j in range(len(slangs)):
  slangs[j]=str(slangs[j].string).lower()
  meanings[j]=str(meanings[j].string).lower()
```

```python
popped=0
for i in range(len(meanings)):
  s=""
  if "online gaming" in meanings[i-popped]:
    slangs.pop(i-popped)
    meanings.pop(i-popped)
    popped+=1
    continue

  elif("\"" in meanings[i-popped]):
    s=re.findall(r'"([^"]*)"', meanings[i-popped])[0]
  else: 
    flag=1
    for j in meanings[i-popped]:
      if j=="(":
        flag=0
      elif j not in "()" and flag==1:
        s=s+j
      else:
        flag=1
    
    meanings[i-popped]=s
```

```python
di={"freak":"fuck","*a*":"ass","*b*":"bitch","*stuff*":"shit","*s":"shit","*":""}
```

```python
for i in range(len(meanings)):
  for j in di:
    meanings[i]=meanings[i].replace(j,di[j])
```

```python
for j in range(len(slangs)-1):
  if (slangs[j][4:]== "Text Message & Chat Abbreviations"):
    continue
  else:
    addToDict(slangs[j],meanings[j])
```

##https://www.socialmediatoday.com/content/top-twitter-abbreviations-you-need-know

```python
#https://www.socialmediatoday.com/content/top-twitter-abbreviations-you-need-know

URL="https://www.socialmediatoday.com/content/top-twitter-abbreviations-you-need-know"
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36'}

page = requests.get(URL,headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')

c = soup.find(class_=' large medium article-body')                             

text=c.find_all("p")
w=str(text[1]).split("\n")
w=w[1:]


t=[i.split("<")[0].split(",")[0].split() for i in w]
q=[[i[1].split("/")[0].lower()," ".join(i[3:]).lower()] for i in t]
q

popped=0
for i in range(len(q)):
  if len(q[i-popped][0])==1 or q[i-popped][0][0]=="#" or q[i-popped][1]=="":
    q.pop(i-popped)
    popped+=1

d={"f---":"fuck","s---":"shit","b----":"bitch"}
for i in q:
  for t in d:
    i[1]=i[1].replace(t, d[t])
```

```python
for i in q:
  addToDict(i[0],i[1])
```
