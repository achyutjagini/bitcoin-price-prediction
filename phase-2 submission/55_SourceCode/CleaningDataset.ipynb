```python
def getInfo(df):
  no_of_cols=df.shape[1]
  inform=[[[]for j in range(5)] for i in range(no_of_cols)]

  attrs=[i for i in df]
  lenght=len(df[attrs[0]])
  for i in range(0,no_of_cols):
    #print(i)
    inform[i][0]=attrs[i]
    inform[i][1]=df[attrs[i]].dtypes
    inform[i][2]=lenght
    inform[i][3]=len(df[attrs[i]].unique())
    inform[i][4]=df[attrs[i]].isna().sum()
    


  inform = pd.DataFrame(inform,columns=['ATTRIBUTE', 'TYPE','VALUES','UNIQUE', 'NANS' ])

  return inform
```
